#  Date Functionality

```python
data = pd.date_range(start = "2022-03-22" , end = "2022-06-22")
data
```

```python
df_date = pd.DataFrame(data, columns = ["date"])
df_date
```

```python
df_date.dtypes
```

```python
df7 = pd.DataFrame(
    {
        "date":["2022-06-22","2022-06-20", "2022-06-19" ]
    }
)
df7
```

```python
df7.dtypes
```

As you can see above, the date during pd.date_range() function had a datatype of datetime64 in dataframe.

But the same date when copied into a list which is a value in a dictionary and then converted into a dataframe, its datatype is string/object.

Hence, if we scrap data from a data which contains date, it can be available in the form of string/object or datetime64

Hence, if the date is in string/object form, we will first convert it into datetime64. This will enable us to perform operations on it

```python
df7["update_date"] = pd.to_datetime(df7["date"])
df7
```

```python
df7.dtypes
```

Hence, you can see the difference in datatypes of dates in date column and that of dates in update_date column

```python
# Create column which stores only year

df7["year"] = df7["update_date"].dt.year
df7
```

```python
# Create column which stores only month

df7["month"] = df7["update_date"].dt.month
df7
```

```python
# Create column which stores only day

df7["day"] = df7["update_date"].dt.day
df7
```

#Python Pandas Time Delta

Time delta is nothing but time difference

```python
pd.Timedelta(days=1)
```

```python
pd.Timedelta(days = 3, hours = 5)
```

```python
pd.Timedelta(days = 2, hours = 18, minutes = 55)
```

```python
dt = pd.to_datetime("2022-06-20")
time = pd.Timedelta(days = 2, hours = 18, minutes = 55)
```

```python
dt + time
```

As you can see, the date has changed from 20 to 22 due to time difference

```python
dt = pd.to_datetime("2022-06-20")
time = pd.Timedelta(days = 1, hours = 5, minutes = 55)

dt + time
```

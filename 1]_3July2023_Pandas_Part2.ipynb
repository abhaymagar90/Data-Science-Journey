Now, we can also use iloc

```python
df.iloc[0:2]   # iloc excludes the  last element of the range
```

```python
df.loc[0:2]  # loc includes the  last element of the range
```

* The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc().

* The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc().

```python
# Select first 2 rows and first 3 columns

df.iloc[0:2, ['PassengerId'	,'Survived'	,'Pclass']]
```

```python
df.iloc[0:2, 0:3]
```

```python
# Select first 2 rows and first 3 columns

df.loc[0:2, ['PassengerId'	,'Survived'	,'Pclass'] ]
```

iloc only works with inbuilt indexes whereas loc works with inbuilt indexes for rows and named indexes for columns.

inbuilt indexes -> 0,1,2,3

default/ named indexes -> PassengerId, Pclass, Sex

```python
df.loc[0:2, 0:3]
```

For loc, slicing can be done for rows and not for columns. This is because for rows, inbuilt index is taken always as there is no named index unlike that of column where named index is present.

```python
# Consider you have to drop some columns from the dataframe

df.drop("PassengerId")
```

We are getting an error ( "['PassengerId'] not found in axis" )

The reason is that, by default drop() operation takes axis = 0

* axis = 0   -> Row
* axis = 1   -> Column

Here rows are indicated by 0,1,2....  and column are represented by PassengerId, Survived, Pclass,...

Hence, to delete a column, we can set axis = 1

```python
df.drop("PassengerId", axis = 1)
```

As you can see, PassengerId column is deleted

```python
df.head()
```

As you can see, df still has PassengerID even if we dropped it. The main reason is that, drop() only drops the column in runtime and not in original dataframe. This is because inplace is by default False

To delete the column in the original dataframe you must set "inplace = True" while usind drop() function

```python
df.drop("PassengerId", axis = 1, inplace = True)
```

```python
df.head()
```

Now you can see that PassengerID is removed from original dataframe

# Difference between Merge and Join

* Join operation works using Row Indexes
* Merge operation works using column values

Both join and merge can be used to combines two dataframes but the join method combines two dataframes on the basis of their indexes whereas the merge method is more versatile and allows us to specify columns beside the index to join on for both dataframes.

 join() is used to combine two DataFrames on the index but not on columns whereas merge() is primarily used to specify the columns you wanted to join on, this also supports joining on indexes and combination of index and columns.

```python
# ADD a new column which will contain Fare in INR

df["Fare_in_INR"] = df["Fare"].apply( lambda x : x*80 )
```

```python
df.head()
```

Hence, in this way you can add a new column and fill values depending upon existing columns

```python
# Add a new column which will contain length of Name

df["Len_Name"] = df["Name"].apply( lambda x : len(x) )
df.head()
```

```python
# Without using lambda function/ using normal function

def inr_conv(x):
  return x*80

df["INR_Fare"] = df["Fare"].apply(inr_conv)
```

```python
df.head()
```

Hence, you can use external function or lambda

```python
# Create a column with 3 categories for fare -> cheap, mid, expensive

def create_flag(x):
  if x<10:
    return "cheap"
  elif x >=10 and x < 20:
    return "mid"
  else:
    return "expensive"
```

```python
df["Flag_Fare"] = df["Fare"].apply(create_flag)
```

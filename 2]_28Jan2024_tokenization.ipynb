```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')

text = "My name is Abhay. I just love data science! Do you feel the same?"
```

```python
# Word Tokenization

word_tokenize(text)
```

```python
# sentence tokenization

sent_tokenize(text)
```

```python
# Using regex

import re

pattern = r" "
removed_special_char = re.sub("[.?!]","", text)
removed_special_char
```

```python
re.split("\s",removed_special_char)
```

# Vertical bar charts are useful to compare different categorical or discrete variables, such as age groups, classes, schools, etc., as long as there are not too many categories to compare.

```python
# Suppose you want to plot a graph where x-axis is data and y-axis is its frequency
# frequency - repetition or occurences of each datapoint

data = [9,9,9,1,2,3,4,5,6,5,6,7,8,8,8,8,8]
```

```python
plt.figure(figsize= (8,5))
plt.hist(data, color = "#C23EE0")
plt.xlabel("Representing unique Value")
plt.ylabel("Represents Frequency")
plt.title("Hist Plot Example")
plt.grid()
plt.savefig("test.png")   # Save Figure in Local Dir
```

When should you use a histogram instead of a bar graph?
The histogram is used to showcase a graphical presentation that represents the data in the form of frequency; whereas a bar chart is also a graphical representation of data and the information that is used for the comparison of two categories. Both bar charts and histograms are an important element of statistics.
